```python
import pandas as pd
import io
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(io.BytesIO(uploaded['wine.csv']))
```

```python
df.head()
```

```python
df.info()
```

```python
# Analyze the shape of the dataset
shape = df.shape
num_rows = shape[0]  # Number of rows
num_columns = shape[1]  # Number of columns

print("Dataset Shape:")
print("Number of rows:", num_rows)
print("Number of columns:", num_columns)

# Analyze the size of the dataset
size = df.size

print("\nDataset Size:")
print("Number of elements:", size)
```

```python
# Compute descriptive statistics
stats = df.describe()
stats
```

```python
# Plot the statistics
sns.set(style='whitegrid')  # Set the style of the plot
plt.figure(figsize=(10, 6))  # Set the size of the plot

# Plotting the statistics using a bar plot
sns.barplot(data=stats, orient='h')

# Customize the plot
plt.title('Descriptive Statistics')
plt.xlabel('Values')
plt.ylabel('Statistics')

# Display the plot
plt.show()
```

```python
# Find the number of null values
null_counts = df.isnull().sum()

# Display the number of null values
print("Number of Null Values:")
null_counts
```

```python
target_column = 'Target'  # Replace 'target' with the actual column name

# Count the number of classes
class_counts = df[target_column].value_counts()

# Display the number of classes
print(f'Number of Classes in {target_column} : {len(class_counts)}')
```

```python
df.Target.unique()
```

```python
cor_mtx = df.corr()
cor_mtx
```

```python
# Plot scatter matrix using seaborn
sns.set(style='ticks')  # Set the style of the plot
sns.pairplot(df)
plt.show()
```

```python
# Plot a heatmap using seaborn
sns.set()
plt.figure(figsize=(10, 8))
sns.heatmap(cor_mtx, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
